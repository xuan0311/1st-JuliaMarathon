# Julia 套件管理 (Package Management)

## Day 003 作業

在接下來的課程中，我們還會用到許多不同套件，請在今天的作業中先行安裝下列套件：`CSV.jl` （將會在資料操作的單元介紹）。

### 指定安裝套件及版本

今日作業請以 Functional API 模式進行套件安裝，並且指定安裝 0.5.14 版。

安裝完成後請呼叫 `Pkg.installed()` 函式檢查已安裝的版本。

啟用套件管理

```julia
using Pkg
```

呼叫 `installed()` 函數查看目前已安裝的套件。

```julia
Pkg.installed()
```

```julia
Pkg.add("CSV")
```

呼叫 `installed()` 函數查看目前已安裝的套件。

```julia
Pkg.installed()
```

刪除安裝好的套件

```julia
Pkg.rm("CSV")
```

安裝指定版本

```julia
Pkg.add(PackageSpec(name="CSV", version="0.5.14"))
```

呼叫 `installed()` 函數查看目前已安裝的套件。

```julia
Pkg.installed()
```

```julia
Pkg.rm("DataFrames")
```

```julia
Pkg.rm("PyCall")
```

呼叫 `installed()` 函數查看目前已安裝的套件。

```julia
Pkg.installed()
```
